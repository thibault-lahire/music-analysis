```python
import os, sys, wave, struct

import numpy as np
import pyaudio
import pandas as pd
import matplotlib.pyplot as plt

from copy import deepcopy
from math import ceil

from scipy.io.wavfile import write
```

```python
def load_music(file):
    return wave.open(file, 'rb')


def play_music(file, chunk = 1024):
    """
    Script from PyAudio doc
    """
    wf = wave.open(file, 'rb')
    p = pyaudio.PyAudio()
    stream = p.open(format=p.get_format_from_width(wf.getsampwidth()),
                    channels=wf.getnchannels(),
                    rate=wf.getframerate(),
                    output=True)
    data = wf.readframes(chunk)

    while data:
        stream.write(data)
        data = wf.readframes(chunk )

    stream.stop_stream()
    stream.close()
    p.terminate()
    
    
def plot_sound(data, times, name='default_name', save=False):
    plt.figure(figsize=(30, 4))
    plt.fill_between(times, data)
    plt.xlim(times[0], times[-1])
    plt.xlabel('time (s)')
    plt.ylabel('amplitude')
    if save:
        plt.savefig(name+'.png', dpi=100)
    plt.show()

def write_file(data, Fs, name):
    scaled = np.int16(data/np.max(np.abs(data)) * 32767)
    write(name+'.wav', Fs, scaled)
    return name+'.wav'

def listen_file(filename):
    current_path = os.getcwd()
    data_path = os.path.join(current_path, 'Sons')
    music = os.path.join(data_path, filename)
    play_music(music)
```

```python
def preprocess_sound(filename='guitare.wav'):
    current_path = os.getcwd()
    data_path = os.path.join(current_path, 'Sons')
    music = os.path.join(data_path, filename)
    wavefile = wave.open(music, 'rb')
    Fs = int(wavefile.getframerate())
    num_samples = int(wavefile.getnframes())
    data = wavefile.readframes(num_samples)
    data = struct.unpack('{n}h'.format(n=num_samples), data)
    x = np.array(data)
    timestep = 1/float(Fs)
    times = np.arange(len(x))*timestep
    return music, x, times, Fs
```

```python
music, x, times, Fs = preprocess_sound()
print("The sampling frequency is {:} Hz".format(Fs))
```

```python
plot_sound(x, times)
play_music(music)
```

# 1. Phasing and 2. Flanger

```python
amin = 0.5 
amax = 1.5 
f_a = 5 # Hz 

A_a = (amax-amin)/2
B_a = (amax+amin)/2

pmax = int(Fs/200)
pmin = int(Fs/400)
f_p = 5 # Hz 

A_p = (pmax-pmin)/2
B_p = (pmax+pmin)/2

def a(n, A_a, B_a, f_a, Fs, phasing=True):
    if phasing:
        return B_a + A_a*np.sin(2*np.pi*f_a*n/Fs)
    else:
        return -1

def p(n, A_p, B_p, f_p, Fs, flanger=True):
    if flanger:
        return int(B_p + A_p*np.sin(2*np.pi*f_p*n/Fs))
    else:
        return 50

def effects(n, x, A_a, B_a, f_a, Fs, A_p, B_p, f_p, phasing=True, flanger=True):
    return x[n] + a(n, A_a, B_a, f_a, Fs, phasing)*x[n-p(n, A_p, B_p, f_p, Fs, flanger)]
```

```python
y = np.zeros_like(x)
for i in range(x.shape[0]-pmax):
    y[i] = effects(i+pmax, x, A_a, B_a, f_a, Fs, A_p, B_p, f_p)

plot_sound(y, times)
write_file(y, Fs, 'Sons/results')
listen_file('results.wav')
```

We now test our implementation on a pure sound of 441 Hz.

```python
s = (np.sin(2*np.pi*441*times)*32767.0/10).astype('int')
plot_sound(s, times)
write_file(s, Fs, 'Sons/sinus441')
listen_file('sinus441.wav')
```

We first do a sanity check : if the oscillation in the phasing and flanger effects are not taken into account, the parameters set of $a$ and $p$ were chosen such that the sinus disappear.

```python
y_check = np.zeros_like(s)
for i in range(y_check.shape[0]-pmax):
    y_check[i] = effects(i+pmax, s, A_a, B_a, f_a, Fs, A_p, B_p, f_p, False, False)

plot_sound(y_check, times)
```

We then test the phasing effect :

```python
y_phasing = np.zeros_like(s)
for i in range(y_phasing.shape[0]-pmax):
    y_phasing[i] = effects(i+pmax, s, A_a, B_a, f_a, Fs, A_p, B_p, f_p, True, False)

plot_sound(y_phasing, times)
write_file(y_phasing, Fs, 'Sons/phasing')
listen_file('phasing.wav')
```

Then we test the flanger effect :

```python
y_flanger = np.zeros_like(s)
for i in range(y_flanger.shape[0]-pmax):
    y_flanger[i] = effects(i+pmax, s, A_a, B_a, f_a, Fs, A_p, B_p, f_p, False, True)

plot_sound(y_flanger, times)
write_file(y_flanger, Fs, 'Sons/flanger')
listen_file('flanger.wav')
```

Now we combine both effects :

```python
y_combine = np.zeros_like(s)
for i in range(y_combine.shape[0]-pmax):
    y_combine[i] = effects(i+pmax, s, A_a, B_a, f_a, Fs, A_p, B_p, f_p, True, True)

plot_sound(y_combine, times)
write_file(y_combine, Fs, 'Sons/combine')
listen_file('flanger.wav')
```

# 3. Artificial reverberation

## 3.1 Early echoes simulation

**What can be done to improve the realism of the model for a real room ?**

To improve the realism, we should take into account all the reverberations, e.g. the reverberations on all the obstables (tables, chairs...), which leads to a very complicated model...

**How many image sources woulf you have if you include second order echoes ?**

To answer this question, one may think of a rubik's cube. The room is the cube at the center you don't see. The first order echoes correspond to the cubes of the center of each face that you see. The second order echoes correspond to all the other cubes of the rubik's cube. Hence you have $9 \times 3 - 1 = 26$ image sources. 

**To your opinion, is it necessary to include higher order echoes ? justify you answer**

It depends on the size of the room and on the intensity of the sound you consider. To decide if higher order echoes are negligeable, consider that the amplitude of the signal is decreasing as $1/r$ with $r$ the distance. 



```python
def output(x, a, p):
    N = len(x)
    y = np.zeros(N)
    for n in range(p, N):
        y[n] = x[n-p] + a*y[n-p]
    return y
```

```python
def early_rev(x, S, M, X, Y, Z, Fs, c):

    D = np.zeros(7)
    x_s = S[0]
    y_s = S[1]
    z_s = S[2]
    x_m = M[0]
    y_m = M[1]
    z_m = M[2]

    # Computation of the distances
    D[0] = np.sqrt((x_s+x_m)**2+(y_s-y_m)**2+(z_s-z_m)**2)
    D[1] = np.sqrt((x_s-x_m)**2+(y_s+y_m)**2+(z_s-z_m)**2)
    D[2] = np.sqrt((x_s-x_m)**2+(y_s-y_m)**2+(z_s+z_m)**2)
    D[3] = np.sqrt((2*X-(x_s+x_m))**2+(y_s-y_m)**2+(z_s-z_m)**2)
    D[4] = np.sqrt((x_s-x_m)**2+(2*Y-(y_s+y_m))**2+(z_s-z_m)**2)
    D[5] = np.sqrt((x_s-x_m)**2+(y_s-y_m)**2+(2*Z-(z_s+z_m))**2)
    D[6] = np.sqrt((x_s-x_m)**2+(y_s-y_m)**2+(z_s-z_m)**2)
    
    y = np.zeros_like(x)

    y_temp = np.zeros_like(x)
    for i in range(7):
        a = min(1/D[i], 1)
        y_temp = output(x, a, int(Fs*D[i]/c))
        y = y + y_temp

    return y
```

```python
c = 340 # sound speed
S = [8,8,3] # location of the source
M = [3,2,4] # location of the microphone
X = 10 # dimension of the room
Y = 10 # dimension of the room
Z = 10 # dimension of the room
y_early = early_rev(x, S, M, X, Y, Z, Fs, c)

write_file(y_early, Fs, 'Sons/early')
listen_file('early.wav')
```

## 3.2 Late reverberation: Schroeder reverberator

For the answer to the questions asked, see the code below. When the room is large, it seems that the music is played in a cathedral !

```python
def comb_filter(x, m, g):
    y = np.zeros(x.shape[0])
    for i in range(len(m)):
        y_temp = output(x, g[i], m[i])
        y = y + y_temp
    return y

def all_pass_filter(x, m, g):
    N = x.shape[0]
    
    y = np.zeros(N)
    for n in range(m[0], N):
        y[n] = x[n-m[0]] + g[0]*(y[n-m[0]] - x[n])

    for i in range(1, len(m)):
        y_temp = np.zeros(N)
        for n in range(m[i], N):
            y_temp[n] = y[n-m[i]] + g[i]*(y_temp[n-m[i]] - y[n])
        y = y_temp

    return y
```

```python
def schroeder(x, Tr, Fs):
    delay_comb = np.array([29.7e-3, 37.1e-3, 41.4e-3, 43.7e-3])
    delay_allpass = np.array([96.83e-3, 32.92e-3])

    Ts = 1/Fs
    m_comb = np.floor(delay_comb/Ts).astype('int')
    m_allpass = np.floor(delay_allpass/Ts).astype('int')

    g_comb = 10**(-3*m_comb*(Ts/Tr))
    g_allpass = np.exp(m_allpass*np.log(1-7*Ts/Tr))
    print("The gains corresponding to the delays given in the TP handout for comb filters are {:}".format(g_comb))
    print("The gains corresponding to the delays given in the TP handout for all pass filters are {:}".format(g_allpass))
    
    y_out_comb = comb_filter(x, m_comb, g_comb)
    y_out = all_pass_filter(y_out_comb, m_allpass, g_allpass)
    return y_out
```

```python
Tr = 0.3 # reverberation time

y_rev = schroeder(x, Tr, Fs)
write_file(y_rev, Fs, 'Sons/reverberation')
listen_file('reverberation.wav')
```
