````python
import os, sys, wave, struct

import numpy as np
import pyaudio
import pandas as pd
import matplotlib.pyplot as plt

from copy import deepcopy
from math import ceil
````

Functions

````python
def load_music(file):
    return wave.open(file, 'rb')


def play_music(file, chunk = 1024):
    """
    Script from PyAudio doc
    """
    wf = load_music(file)
    p = pyaudio.PyAudio()
    stream = p.open(format=p.get_format_from_width(wf.getsampwidth()),
                    channels=wf.getnchannels(),
                    rate=wf.getframerate(),
                    output=True)
    data = wf.readframes(chunk)

    while data:
        stream.write(data)
        data = wf.readframes(chunk )

    stream.stop_stream()
    stream.close()
    p.terminate()
    
    
def plot_sound(data, times, name='default_name', save=False):
    plt.figure(figsize=(30, 4))
    plt.fill_between(times, data)
    plt.xlim(times[0], times[-1])
    plt.xlabel('time (s)')
    plt.ylabel('amplitude')
    if save:
        plt.savefig(name+'.png', dpi=100)
    plt.show()
    

def nextpow2(x):
    assert x>0
    p = ceil(np.log2(x))
    x_ = 2**p
    assert 2**(p-1) < x <= x_
    return p, x_


def f2idx(F, df):
    """
    Convert frequency to corresponding index in "frequencies" array
    """
    return ceil(F/df)
````

# I- Windowing and Fourier transform

All the computational steps provided in the template in order to help us are summarized in the cell below

````python
def preprocess_sound(filename='A4_piano.wav', plot=True):
    current_path = os.getcwd()
    data_path = os.path.join(current_path, 'sons_multipitch')
    music = os.path.join(data_path, filename) 
    wavefile = load_music(music)
    Fs = int(wavefile.getframerate())
    num_samples = int(wavefile.getnframes())
    data = wavefile.readframes(num_samples)
    data = struct.unpack('{n}h'.format(n=num_samples), data)
    x = np.array(data)
    timestep = 1/float(Fs)
    times = np.arange(len(x))*timestep
    N=ceil(0.7*Fs)       # Window size of analysed signal (only one window of signal is analysed)
    dF_min=Fs/N   # Minimal frequency resolution
    w=np.hamming(N)  # Window
    width = 4*dF_min # largeur du pic spectral (en Hz) 4*dF_min
    eps=float(1e-20)   #precision
    offset=ceil(0.1*Fs)
    xw=x[offset:offset+N]*w    # xw is the analysed signal frame
    n_xw = xw.size
    #Minimal number of data points to satisfy the minimal frequency resolution
    Nfft_min=Fs/dF_min
    #compute the smallest power of two that satisfies the minimal frequency resolution for FFT
    p, Nfft = nextpow2(Nfft_min)
    x_fft = np.fft.fft(xw, n=Nfft) #calcul FFT
    x_fft /= np.max(np.abs(x_fft))+eps # Normalization
    df=Fs/Nfft # frequency virtual resolution of FFT 
    frequencies = np.arange(Nfft)*df  # equivalent to np.fft.fftfreq(n=Nfft, d=timestep)
    X_fft = np.abs(x_fft) # absolute value of FFT (phase doesn't play a role)
    if plot:
        plt.figure(figsize=(30, 4))
        plt.plot(frequencies[:int(Nfft/2)], 20*np.log10(X_fft[:int(Nfft/2)]), color='k')
        plt.xlim(0, 5000)
        plt.ylim(-80, 0)
        plt.xlabel('FrÃ©quences (Hz)')
        plt.ylabel('Spectre d\'Amplitude (dB)')
        plt.title('Spectre d\'amplitude du signal original')
        plt.show()
    return frequencies, X_fft, Nfft, width, dF_min
````

## II.2 - Computation of the spectral sum 

````python
Fmin=100             # Minimal F0 frequency that can be detected
Fmax=900             # Maximal F0 frequency that can be detected
H=4                  # H = nombre de versions compressÃ©es
Rmax = ceil((Nfft-1)/(2*H)) # frÃ©quence maximale
print("Because Nfft has a value of", Nfft)
print("Then the maximum value of R that can be computed is", Rmax)
print("And the corresponding frequency is {:} Hz".format(freq[Rmax]))
````

Indeed, note that $F_s = 32000$ Hz. This means that all information of interest is contained from frequency 0 Hz up to $F_s/2 = 16000$ Hz. By definition of the spectral sum, if we go to order H, we divide the interval of interest by H. Hence we have $R_{max,freq} = F_s / (2*H)$.

````python
def concentrate_spectrum(x, h, R):
    '''
    Function that takes as input a vector x, i.e. the module of the fourier transformation,
    a integer h and a maximum length R, and outputs a vector of size R that encodes the module of the fourier
    transformation multiplied by h
    '''
    assert len(x)>h*R, "input array is not long enough"
    h = int(h)
    res = np.zeros(R)
    for i in range(R):
        for k in range(h):
            res[i] += x[h*i+k]
        res[i] = res[i]*1.0/h
    return res
````

## II.3 - Finding the spectral maximum

In the cells below, we propose an implementation of the spectral sum and the spectral product. Change the boolean parameter ```product``` to test one method or the other. As asked in the TP handout, I have tested the two techniques on A4_piano and E4_oboe and did not notice any difference.

````python
R = 4000    # Rmax = 4096
product = False
````

````python
def spectral_computation(fourier, H, R, product=False):
    '''Computation of the spectral sum if product==False, else computation of the spectral product'''
    
    modules = np.zeros((H, R))
    for h in range(1,H+1):
        modules[h-1] = concentrate_spectrum(fourier, h, R)
    if product:
        return np.prod(modules, axis=0)
    else:
        return np.sum(modules, axis=0)


def get_fundamental(freq, spec_computation, H, R, Fmin, Fmax, product=False, plot=True):
    '''Returns the fundamental frequency and the index of this fundamental frequency'''
    
    if plot:
        plt.figure(figsize=(30, 4))
        plt.plot(freq[:R], spec_computation, color='k')
        plt.xlim(Fmin, Fmax)
        #plt.ylim(-80, 0)
        plt.xlabel('Frequencies (Hz)')
        plt.ylabel('Amplitude spectrum (dB)')
        plt.title('Spectral sum')
        plt.show()
    
    ind_min = len(freq[freq<Fmin])
    ind_max = len(freq[freq<Fmax])
    
    if product:
        return freq[np.argmin(spec_computation[ind_min:ind_max])+ind_min], np.argmin(spec_computation[ind_min:ind_max])+ind_min
    else:
        return freq[np.argmax(spec_computation[ind_min:ind_max])+ind_min], np.argmax(spec_computation[ind_min:ind_max])+ind_min
````

````python
f_0, idx_f_0 = get_fundamental(freq, spectral_computation(fourier, H, R, product), H, R, Fmin, Fmax, product)
print("The fundamental frequency is {:.5} Â± {:.3} Hz".format(f_0, dF_min))
````

# III - Subtraction of the sound corresponding to the detected fundamental frequency

## III.1 - Harmonics detection

The search of the k-th harmonic around the frequency $f_k = k.F_0$ can be inappropriate for music instruments "suffering" from inharmonicity, such as piano. However, if we take the example of oboe, it works very well. 

The search is done between $f_{kmin} = (1 - \alpha)*f_k$ and $f_{kmax} = (1 + \alpha)*f_k$. A good choice for $\alpha$ is $2^{1/12}$. Indeed, music instruments are supposed to play notes that are separated by at least a semitone. If frequency $x$ can be played, then the musician can also play frequency $x*2^{1/12}$ and $x/2^{1/12}$, but nothing in between.

## III.2 - Harmonics suppression

In what follows, the pics of the harmonics are detected directly on the module of the Fourier transform, and we use the variable  ```width``` to "cut" them. I have also tested another method. Starting from the pic, I explore the points next to the pic. While the energy is decreasing, I consider the point as part of the pic. When the energy begins to oscillate, I consider the point as out of the pic. This way, I find another width of the pic. In practice, I did not notice a significant difference...

````python
def delete_harmonic(freq, fourier, idx_note, width, plot=True):
    
    gamma_plus = idx_note + len(freq[freq<width]) + 2
    gamma_moins = idx_note - len(freq[freq<width]) - 2

    if fourier[gamma_plus]<fourier[gamma_moins]:
        gamma = gamma_plus - idx_note
        freq_in_gamma = fourier[gamma_plus]
    else:
        gamma = idx_note - gamma_moins
        freq_in_gamma = fourier[gamma_moins]

    count = idx_note - gamma
    while count < idx_note + gamma:
        fourier[count] = freq_in_gamma
        count += 1
    
    if plot:
        plt.figure(figsize=(30, 4))
        plt.plot(freq, fourier, color='k')
        plt.xlim(0, 5000)
        plt.ylim(-80, 0)
        plt.xlabel('FrÃ©quences (Hz)')
        plt.ylabel('Spectre d\'Amplitude (dB)')
        plt.title('Modified spectrum')
        plt.show()    
    
    return freq, fourier
````

The next step of our implementation is to detect the harmonics and suppress them. However, as we wrote, we might encounter inharmonicity, as is the case for piano. In such a case, the search interval $[f_{kmin},f_{kmax}]$ for each harmonic has to be changed by defining $f_{kmin} = (1-\alpha)*f_{inharmo}$ and $f_{kmax} = (1+\alpha)*f_{inharmo}$ with $f_{inharmo} = k*F_0*\sqrt{(1+k^2 \beta)}$. To be more coherent, we write $f_1 = F_0$

This means that $f_k \simeq k*f_1*\sqrt{(1+k^2 \beta)}$, $\quad$ i.e. $\quad$  $f_k^2 \simeq k^2*f_1^2*(1+k^2 \beta)$, $\quad$ i.e. $\quad$ $\beta \simeq \frac{ \left( \frac{f_k}{k} \right)^2 - f_1^2 }{(k f_1)^2}$


We use this calculation to propose an automatic way to find the coefficient $\beta$. We compute $\beta$ for the first harmonics (up to 2500 Hz) and then we average the different values of $\beta$ found.

````python
def find_beta(freq, fourier, idx_note, freq_funda):
    freq_study = 2*freq_funda
    count = 2
    semitone_above = freq_funda*2**(1/12)
    alpha = semitone_above - freq_funda
    alpha_idx = len(freq[freq<alpha])
    current_idx_array = []
    
    while freq_study < 2500:
        current_idx = np.argmax(fourier[count*idx_note-alpha_idx:count*idx_note+alpha_idx]) + count*idx_note - alpha_idx
        current_idx_array.append(current_idx)
        count += 1
        freq_study = count*freq_funda
    
    current_idx_array = np.array(current_idx_array)
    associated_freq = freq[current_idx_array]
    beta_candidates = np.zeros_like(associated_freq)
    for i in range(beta_candidates.shape[0]):
        beta_candidates[i] = ((associated_freq[i]/(i+2))**2 - freq_funda**2)/((i+2)*freq_funda)**2
    beta = np.mean(beta_candidates)
    return beta

beta = find_beta(fr, fo, idx_f_0, f_0)
print("beta = ", beta)
````

````python
def delete_note(freq, fourier, idx_note, freq_funda, width, inharmonicity=False, plot=True):
    count = 2
    if inharmonicity:
        beta = find_beta(freq, fourier, idx_note, freq_funda)
        freq_study = count*freq_funda*np.sqrt(np.abs(1+beta*count**2))
    else:
        freq_study = count*freq_funda
    center_idx = ceil(freq_study/(freq[1]-freq[0]))
    
    semitone_above = freq_funda*2**(1/12)
    alpha = semitone_above - freq_funda
    alpha_idx = 2*len(freq[freq<alpha])
    
    while freq_study < 15000 and center_idx < len(freq):
        if center_idx-alpha_idx<0:
            current_idx = np.argmax(fourier[0:center_idx+alpha_idx])
        else:
            current_idx = np.argmax(fourier[center_idx-alpha_idx:center_idx+alpha_idx]) + center_idx-alpha_idx
        freq, fourier = delete_harmonic(freq, fourier, current_idx, width, plot)
        
        count += 1
        if inharmonicity:
            freq_study = count*freq_funda*np.sqrt(np.abs(1+beta*count**2))
        else:
            freq_study = count*freq_funda 
        center_idx = ceil(freq_study/(freq[1]-freq[0]))
        
    return freq, fourier
````

## III.3 - Stopping criterion

We now put all together what we've done before. Note that the following cells are independent from the rest of the notebook. To test on different sounds, you just have to modify the cells below. 

The idea of this last part is to tackle the issue of multipitch detection. A sound is given, we preprocess it, and we want to find the different fundamental frequencies it is made of. 

The main issue is to find a good stopping criterion: when do we know that the algorithm has found all the fundamental frequencies ? I chose to check if the amplitude spectrum was below a certain value (-25 dB) in the interval [Fmin,Fmax]. By checking this, I make sure there is no pic remaining. 

````python
def note_detection(filename, plot=False):
    frequencies, X_fft, Nfft, width, _ = preprocess_sound(filename)
    freq = frequencies[:int(Nfft/2)] 
    fourier = 20*np.log10(X_fft[:int(Nfft/2)])

    Fmin = 100    # Minimal F0 frequency that can be detected
    Fmax = 900    # Maximal F0 frequency that can be detected
    H = 4
    R = 4000      # Rmax = 4096
    
    ind_min = len(freq[freq<Fmin])
    ind_max = len(freq[freq<Fmax])
    
    product = False
    inharmonicity = True
    
    notes_array = []
    
    while np.max(fourier[ind_min:ind_max]) > -25: # Stopping criterion
        f_0, idx_f_0 = get_fundamental(freq, spectral_computation(fourier, H, R, product), H, R, Fmin, Fmax, product, plot)
        notes_array.append(f_0)
        print("Current fundamental frequencies detected:", notes_array)
        fr, fo = delete_harmonic(freq, fourier, idx_f_0, width, plot)
        freq, fourier = delete_note(fr, fo, idx_f_0, f_0, width, inharmonicity, plot)

        plt.figure(figsize=(30, 4))
        plt.plot(freq, fourier, color='k')
        plt.xlim(0, 5000)
        plt.ylim(-80, 0)
        plt.xlabel('FrÃ©quences (Hz)')
        plt.ylabel('Spectre d\'Amplitude (dB)')
        plt.title('Modified spectrum')
        plt.show()
        
        
    return notes_array
````
